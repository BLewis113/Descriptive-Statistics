```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
weibull = np.random.weibull(7, 100)

plt.hist(weibull)
plt.show()
```

```python
plt.hist(weibull)

plt.axvline(weibull.mean(), color = 'r', linestyle = 'solid', linewidth = 2)
plt.axvline(weibull.mean() + weibull.std(), color = 'r', linestyle = 'dashed', linewidth =2)
plt.axvline(weibull.mean() - weibull.std(), color = 'r', linestyle = 'dashed', linewidth =2) 

def values(weibull):
    print("mean: ",np.mean(weibull))
    print("std. deviation :",np.std(weibull))

values(weibull)
```
